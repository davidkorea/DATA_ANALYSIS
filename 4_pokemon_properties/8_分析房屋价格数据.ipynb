```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. 通过盒形图可视化不同卧室个数对应的房屋价格的分布 
2. 通过双变量图观察卫生间个数与房屋价格的关系 
3. 通过热图可视化变量间的关系 

```python
data_df = pd.read_csv('../data/house_data_full.csv')
data_df.head()
```

```python
data_df.isnull().sum(axis=0)
```

```python
data_df.info()
```

```python
data_df_corr = data_df.corr()
```

```python
plt.figure(figsize=(15,9))
sns.heatmap(data_df_corr, cmap='coolwarm', annot=True)
```

```python
plt.figure(figsize=(9,6))
sns.boxplot(x='bedrooms', y='price', data=data_df)
plt.tight_layout()
```

```python
sns.jointplot(x='bathrooms', y='price', data=data_df)
plt.tight_layout()
```

```python
plt.figure(figsize=(9,6))
plt.scatter(x='bathrooms', y='price', data=data_df)
plt.tight_layout()
```
